```python
import pandas as pd
```

```python
df=pd.read_csv("nasser.csv")
```

```python
df1=df.dropna(axis=0)
```

```python
genre_counts = df["Genre"].value_counts()
most_popular_genre = genre_counts.idxmax()
print("The most popular genre is:", most_popular_genre)
```

```python
genre_counts = df.groupby("Genre").size().reset_index(name="count")
df1
```

```python
import matplotlib.pyplot as plt


plt.bar(genre_counts['Genre'], genre_counts['count'])
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Most Popular Genre of Movies/TV Shows')
plt.xticks(rotation=90)
plt.show()

```

```python
threshold = 8.0  # Set the threshold for the minimum rating
top_rated_df = df[df["IMDB_Rating"] >= threshold]
```

```python
director_counts = top_rated_df["Director"].value_counts().reset_index()
director_counts.columns = ["Director", "Count"]
director_counts
```

```python
director_counts.plot(kind="bar")
plt.show()
```
